# 1. Import library

```python
from selenium import webdriver
import time
```

# 2. Create a chrome webdriver

```python
driver = webdriver.Chrome()
```

- By default, the browser will `openup as small window`

```python
# To show the demo of transition to the maximum window

driver = webdriver.Chrome()

time.sleep(5)
driver.maximize_window()
```

# 3. How to open any Url ?

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)
```

# 4. How to get the webpage details ?

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)

print(f"Title : {driver.title}")
print(f"URL   : {driver.current_url}")
```

# 5. How to take the webpage screenshot ?

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)

print(f"Title : {driver.title}")
print(f"URL   : {driver.current_url}")

driver.save_screenshot("google-screenshot.png")
print("\nscreenshot taken :)")
```

# 6. How to close the browser ?

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)

print(f"Title : {driver.title}")
print(f"URL   : {driver.current_url}")

driver.save_screenshot("google-screenshot.png")
print("\nscreenshot taken :)")

time.sleep(5)

# It closes the browser automatically
driver.quit()
print("\nBrowser closed successfully :)")
```

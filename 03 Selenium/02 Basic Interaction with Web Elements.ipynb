# 3. Find the Search bar Xpath and use `send_keys` function

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)
time.sleep(2)

# To find the Xpath of serach bar
search_bar_xpath ='//*[@id="APjFqb"]'
search_bar = driver.find_element(by = By.XPATH, value = search_bar_xpath)

# To type text input into search bar
search_bar.send_keys('Machine Learning')
time.sleep(1)
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = 'https://demoqa.com/select-menu'
driver.get(url)
time.sleep(2)

cars_element = driver.find_element(by = By.XPATH, value = '//*[@id="cars"]')
cars_ms = Select(cars_element)
time.sleep(2)

cars_ms.select_by_index(1)
time.sleep(2)

cars_ms.select_by_visible_text("Opel")
time.sleep(2)

cars_ms.select_by_visible_text("Audi")
time.sleep(2)

cars_ms.deselect_by_index(2)
time.sleep(2)

cars_ms.deselect_all()
time.sleep(2)

driver.quit()
```

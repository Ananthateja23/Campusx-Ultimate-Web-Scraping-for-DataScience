## Working on dropdown

````python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
````

````python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.miniclip.com/careers/vacancies"
driver.get(url)
time.sleep(2)

locations_field = driver.find_element(by = By.XPATH, value = '//fieldset[2]//select[1]')
# driver.execute_script("window.scrollBy(0, 250);")
location_dropdown = Select(locations_field)
time.sleep(3)

location_dropdown.select_by_index(1)
time.sleep(3)

location_dropdown.select_by_index(0)
time.sleep(3)

location_dropdown.select_by_visible_text("London, United Kingdom")
time.sleep(3)

driver.quit()
````

# 03 Advanced Web Interactions

````python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
````

````python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)
time.sleep(2)

search_bar = driver.find_element(by = By.ID, value = "APjFqb")
search_bar.send_keys("Machine Learning")

wait = WebDriverWait(driver, 5)
wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@class='FPdoLc lJ9FBc']//input[@name='btnK']")))

search_bar.send_keys(Keys.ENTER)
````

````python
driver = webdriver.Chrome()
driver.maximize_window()

driver.implicitly_wait(5)

url = "https://www.google.com"
driver.get(url)
time.sleep(2)

search_bar = driver.find_element(by = By.ID, value = "APjFqb")
search_bar.send_keys("Machine Learning")
time.sleep(2)

search_bar.send_keys(Keys.ENTER)
time.sleep(2)

driver.quit()
````

## Working on IFrames

````python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/html/tryit.asp?filename=tryhtml_iframe_target"
driver.get(url)
time.sleep(2)

# switch context to frame
iframe_element = driver.find_element(by = By.XPATH, value = '//*[@id="iframeResult"]')
driver.switch_to.frame(iframe_element)
time.sleep(2)

link = driver.find_element(by = By.XPATH, value = '/html/body/p[1]/a')
link.click()
time.sleep(2)

# switch back to main content
driver.switch_to.default_content()
driver.quit()
````

## Woking on Page Object Model(POM)

````python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
````

````python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://github.com/login"
driver.get(url)
time.sleep(2)

driver.implicitly_wait(5)

class Login:
    def __init__(self, driver):
        self.driver = driver
        self.username = (By.ID, "login_field")
        self.password = (By.ID, "password")
        self.login_button = (By.XPATH, "//input[@name='commit']")

    def login(self, username, password):
        self.driver.find_element(*self.username).send_keys(username)
        self.driver.find_element(*self.password).send_keys(password)
        time.sleep(2)
        self.driver.find_element(*self.login_button).click()


login_page = Login(driver)
login_page.login("your username", "your password")
time.sleep(2)

driver.quit()
````

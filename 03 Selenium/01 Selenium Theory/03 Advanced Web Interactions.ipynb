# 1. Explicit Waits

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)
time.sleep(2)

search_bar_xpath = '//*[@id="APjFqb"]'
search_bar = driver.find_element(by = By.XPATH, value = search_bar_xpath)
search_bar.send_keys("Machine Learning")
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)
time.sleep(2)

search_bar_xpath = '//*[@id="APjFqb"]'
search_bar = driver.find_element(by = By.XPATH, value = search_bar_xpath)
search_bar.send_keys("Machine Learning")

wait = WebDriverWait(driver, 5)
search_button_xpath = '/html/body/div[2]/div[4]/form/div[1]/div[1]/div[3]/center/input[1]'
wait.until(EC.element_to_be_clickable((By.XPATH, search_button_xpath)))

search_bar.send_keys(Keys.ENTER)
```

# 2. Implicit Waits

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Edge()
driver.maximize_window()

driver.implicitly_wait(10)

url = "https://www.google.com"
driver.get(url)
time.sleep(1)

search_bar_xpath = '//*[@id="APjFqb"]'
search_bar = driver.find_element(by = By.XPATH, value = search_bar_xpath)
search_bar.send_keys("Machine Learning")
time.sleep(2)

search_bar.send_keys(Keys.ENTER)
time.sleep(2)

driver.quit()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/html/tryit.asp?filename=tryhtml_iframe_target"
driver.get(url)
time.sleep(2)

# switch context to frame
iframe_element = driver.find_element(by = By.XPATH, value = '//*[@id="iframeResult"]')
driver.switch_to.frame(iframe_element)
time.sleep(2)

link = driver.find_element(by = By.XPATH, value = '/html/body/p[1]/a')
link.click()
time.sleep(2)

# switch back to main content
driver.switch_to.default_content()
driver.quit()
```

# 4. Handling alerts

- ## Javascript alerts

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/js/tryit.asp?filename=tryjs_alert"
driver.get(url)
time.sleep(2)

iframe_element = driver.find_element(by = By.ID, value = 'iframeResult')
driver.switch_to.frame(iframe_element)

button = driver.find_element(by = By.XPATH, value = '/html/body/button')
button.click()
time.sleep(2)

print(f"Alert Text: {driver.switch_to.alert.text}")
driver.switch_to.alert.accept()
time.sleep(2)

driver.switch_to.default_content()
driver.quit()
```

- ## Prompt alerts

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/js/tryit.asp?filename=tryjs_prompt"
driver.get(url)
time.sleep(2)

iframe_element = driver.find_element(by = By.ID, value = 'iframeResult')
driver.switch_to.frame(iframe_element)

button = driver.find_element(by = By.XPATH, value = '/html/body/button')
button.click()
time.sleep(2)

print(f"Alert Text: {driver.switch_to.alert.text}")
driver.switch_to.alert.send_keys("are you ready???")
time.sleep(3)

driver.switch_to.alert.accept()
time.sleep(2)

driver.switch_to.default_content()
driver.quit()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/js/tryit.asp?filename=tryjs_prompt"
driver.get(url)
time.sleep(2)

iframe_element = driver.find_element(by = By.ID, value = 'iframeResult')
driver.switch_to.frame(iframe_element)

button = driver.find_element(by = By.XPATH, value = '/html/body/button')
button.click()
time.sleep(2)

print(f"Alert Text: {driver.switch_to.alert.text}")
driver.switch_to.alert.send_keys("are you ready???")
time.sleep(3)

# To click the ok button
driver.switch_to.alert.accept()
time.sleep(2)

button = driver.find_element(by = By.XPATH, value = '/html/body/button')
button.click()
time.sleep(2)

# To click the cancel button
driver.switch_to.alert.dismiss()
time.sleep(2)

driver.switch_to.default_content()
driver.quit()
```

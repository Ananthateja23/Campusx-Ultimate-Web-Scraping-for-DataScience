# Import libraries

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

# Using the `input` function

```python
try:
    driver = webdriver.Edge()
    driver.maximize_window()
    
    url = 'https://www.hackthissite.org/user/login'
    driver.get(url)
    time.sleep(2)
    
    # To find the username, password and login webpage elements
    username_field = driver.find_element(by = By.ID, value = "login_username")
    password_field = driver.find_element(by = By.ID, value = "login_password")
    login_bt_xpath = '/html/body/table/tbody/tr[2]/td/table/tbody/tr/td[2]/form/table/tbody/tr[4]/td/input'
    login_button = driver.find_element(by = By.XPATH, value = login_bt_xpath)
    
    username_field.send_keys("abc") # give any username
    password_field.send_keys("123") # give any password
    login_button.click()
    
    # To find the captcha element
    captcha_xpath = '/html/body/table/tbody/tr[2]/td/table/tbody/tr/td[2]/form/table/tbody/tr[5]/td/img'
    captcha_element = driver.find_element(by = By.XPATH, value = captcha_xpath)
    
    if captcha_element:
        x = input("It will halt the script.. solve your captcha..")
except Exception as e:
    print("Inside Except block", end="\n\n")
    print(e)
else:
    print("\nCaptcha handled! write rest of the script..")
    time.sleep(2)
finally:
    driver.quit()
```

```python
player_names = []
team_names = []
goals = []
assists = []
num_matches = []
shots = []
```

```python
try:
    url = "https://www.bbc.com/sport/football/premier-league/top-scorers"
    response = requests.get(url)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    soup = BeautifulSoup(response.content, "html.parser")
    players = soup.find("tbody").find_all("tr", class_ = "ssrcss-hpa5ss-TableRowBody e1icz100")
    for player in players:

        # extract the attribute values
        player_name  = player.find("div", class_ = "ssrcss-1rqvnx-PlayerName e1n8xy5b1").get_text(strip = True)
        team_name    = player.find("div", class_ = "ssrcss-cfavl3-TeamsSummary e1n8xy5b0").get_text(strip = True)
        goals_scored = int(player.find("div", class_="ssrcss-1fyor78-CellWrapper ef9ipf0").get_text(strip = True))
        stats = player.find_all("div", class_ = "ssrcss-ero735-CellWrapper ef9ipf0")
        assists_made = int(stats[0].get_text(strip = True))
        matches_played = int(stats[2].get_text(strip = True))
        shots_taken = int(stats[-3].get_text(strip = True))

        # store the attribute values into lists
        player_names.append(player_name)
        team_names.append(team_name)
        goals.append(goals_scored)
        assists.append(assists_made)
        num_matches.append(matches_played)
        shots.append(shots_taken)

    # create a dictionary
    data = {
        "players": player_names,
        "teams": team_names,
        "goals": goals,
        "assists": assists,
        "matches": num_matches,
        "shots": shots
    }
    
    # create a dataframe
    df_players = pd.DataFrame(data)
```

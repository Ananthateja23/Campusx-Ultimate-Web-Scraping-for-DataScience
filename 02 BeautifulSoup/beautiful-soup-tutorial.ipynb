```python
# children: It gives the immediate child tags of the current working tag
```

```python
soup.body
```

```python
soup.body.children
```

```python
for child in soup.body.children:
    print(child)
```

```python
soup.head
```

```python
soup.head.children
```

```python
for child in soup.head.children:
    print(child)
```

```python
# descendants: It provides the immediate child tags and their nested child tags of the immediate child tag of the current working tag, recursively
```

```python
soup.body
```

```python
soup.body.descendants
```

```python
for descendant in soup.body.descendants:
    print(descendant)
```

- you make sure to compare the results of `children` and `descendant`

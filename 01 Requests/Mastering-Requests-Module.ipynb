# 4. Working with `POST` Method

```python
uri = "https://httpbin.org/post"

data = {
    "username": "Emilia Clarke",
    "password": "Daenerys Targaryen"
}

response = requests.post(uri, data = data)
```

```python
response.status_code
```

```python
response.json()
```

- By default, the data we send using requests is transmitted as form data.
- In headers, the content type value `x-www-form-urlencoded` indicates the data format sent in the request.

```python
response.json()['headers']
```

```python
response.json()['headers']['Content-Type']
```

```python
response.json()['headers']['Content-Type'].split("/")[-1]
```

```python
# function that gives us data format

def get_data_format(response_got):
    json_response = response_got.json()
    data_format = json_response["headers"]["Content-Type"].split("/")[-1]
    print(f"Response data format: {data_format}")
```

```python
get_data_format(response)
```

# 7. Working with HTTP Headers
- We use a GET request to work with headers
- Url: `https://httpbin.org/headers`

```python
headers = {
	'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
    'Accept': 'application/json',
	'Authorization': 'Bearer YOUR_ACCESS_TOKEN',
	'Content-Type': 'application/json',
	'X-Custom-Header': 'CustomValue',
}

uri = "https://httpbin.org/headers"
```

```python
response = requests.get(uri, headers = headers)
```

# 9. Working with a public API
- API: `https://jsonplaceholder.typicode.com/`
- We use the endpoint: `posts`
- To perform the error handling, use `raise_for_status()`
- If no error exists, then `raise_for_status()` gives `None`

```python
url = "https://jsonplaceholder.typicode.com/posts"
```

## 9.1
- To make a GET request to retrieve a list of posts

```python
try:
	response = requests.get(url)
	response.raise_for_status()
except Exception as e:
	print(e)
else:
	status_code = response.status_code
	print(f"Status Code: {status_code}")
	if status_code == 200:
		print("\nSuccessful GET request!")
		posts = response.json()
		for i in range(3):
			print(f"\nPost {i + 1}:")
			print(posts[i])
	else:
		print("Unsuccessful GET request!")
```

## 9.2
- To make a POST request to submit a new post
- We include the following parameters:
    - title
    - body
    - userId

```python
new_post = {
	"title": "Sample Post",
	"body": "This is a sample post",
	"userId": 101
}
```

```python
try:
	response = requests.post(url, data=new_post)
	response.raise_for_status()
except Exception as e:
	print(e)
else:
	status_code = response.status_code
	if status_code == 201:
		print(f"Status Code: {status_code}")
		print("\nSuccessful POST request!")
		post = response.json()
		print("\nPost:")
		print(post)
	else:
		print(f"Status Code: {status_code}")
		print("\nUnsuccessful POST request!")
```
